## 5. Fraud percentage by profession

```python
job_counts = master_df["job"].value_counts()

valid_jobs = job_counts[job_counts >= 5000].index

filtered_job_fraud_rate = (
    master_df[master_df["job"].isin(valid_jobs)]
    .groupby("job")["is_fraud"]
    .mean() * 100
)
print(
    filtered_job_fraud_rate
    .sort_values(ascending = False)
    .head(15)
)
```

## 6. Visualize filtered fraud percentages

```python
plt.figure(figsize = (12,5))

filtered_job_fraud_rate.sort_values(ascending = False).head(10).plot(kind = "bar", color = "orange")

plt.title("Fraud Percentage by Profession (Filtered)")
plt.xlabel("Profession")
plt.ylabel("Fraud Percentage")

plt.xticks(rotation = 45)

plt.show()
```

## Correlation Heatmap

## 1. Select numerical columns

```python
numerical_cols = [
    "amt",
    "city_pop",
    "lat",
    "long",
    "merch_lat",
    "merch_long",
    "unix_time",
    "customer_age",
    "transaction_hour",
    "is_fraud"
]
```

## 2. Create correlation matrix

```python
correlation_matrix = (
    master_df[numerical_cols]
    .corr()
)
correlation_matrix
```

## 3. Plot heatmap

```python
plt.figure(figsize = (12,8))

sns.heatmap(
    correlation_matrix,
    annot = True,
    cmap = "coolwarm",
    fmt = ".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```
