 ## Create tables using sample data

## Load only 1000 rows first of transactions_master

```python
sample_transactions = transactions_df.head(1000)
```

```python
sample_transactions.to_sql(
    "transactions_master",
    con=engine,
    if_exists="replace",
    index=False
)

print("Table Created Successfully")
```

## Load data of fraud_risk_scores

```python
risk_df.to_sql(
    "fraud_risk_scores",
    con=engine,
    if_exists="replace",
    index=False
)

print("Risk Table Created Successfully")
```

```python
chunk_size = 50000

for i, chunk_start in enumerate(
    range(0, len(transactions_df), chunk_size)
):
    
    chunk = transactions_df.iloc[
        chunk_start : chunk_start + chunk_size
    ]
    
    if i == 0:
        
        chunk.to_sql(
            "transactions_master",
            con=engine,
            if_exists="replace",
            index=False
        )
        
    else:
        
        chunk.to_sql(
            "transactions_master",
            con=engine,
            if_exists="append",
            index=False
        )
    
    print(
        f"Loaded rows: {chunk_start + len(chunk):,}"
    )

print("Full Dataset Loaded Successfully")
```
