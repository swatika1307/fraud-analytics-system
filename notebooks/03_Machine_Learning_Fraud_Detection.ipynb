## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load datasets

```python
X_train_smote = pd.read_csv("../datasets/X_train_smote.csv")

X_test = pd.read_csv("../datasets/X_test.csv")

y_train_smote = pd.read_csv("../datasets/y_train_smote.csv")

y_test = pd.read_csv("../datasets/y_test.csv")
```

## Check shapes

```python
print(X_train_smote.shape)

print(X_test.shape)

print(y_train_smote.shape)

print(y_test.shape)
```

```python
# Converting DataFrames to Series

y_train_smote = y_train_smote.squeeze()
y_test = y_test.squeeze()
```

```python
print(type(y_train_smote))

print(type(y_test))
```

## Train First Machine Learning Model

## 1. Import model

```python
from sklearn.linear_model import LogisticRegression
```

## 2. Train model

```python
log_model = LogisticRegression(max_iter = 1000)

log_model.fit(X_train_smote, y_train_smote)
```

## 3. Generate the predictions

```python
y_pred_log = log_model.predict(X_test)
```

## 4. Check prediction sample

```python
print(y_pred_log[:20])
```

## Evaluate Logistic Regression Model

## 1. Import metrics

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score
)
```

## 2. Accuracy

```python
log_accuracy = accuracy_score(y_test, y_pred_log)

print("Accuracy: ", log_accuracy)
```

## 3. Precision

```python
log_precision = precision_score(y_test, y_pred_log)

print("Precision: ", log_precision)
```

## 4. Recall 

```python
log_recall = recall_score(y_test, y_pred_log)

print("Recall: ", log_recall)
```

## 5. F1 Score

```python
log_f1 = f1_score(y_test, y_pred_log)

print("F1 Score: ", log_f1)
```

## 6. ROC-AUC Score

```python
log_roc_auc = roc_auc_score(y_test, y_pred_log)

print("ROC-AUC Score: ", log_roc_auc)
```

## 7. Confusion Matrix

```python
log_cm = confusion_matrix(y_test, y_pred_log)

print(log_cm)
```

## 8. Classification Report

```python
print(classification_report(y_test, y_pred_log))
```
